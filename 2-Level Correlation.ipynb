```python
correlations = []
Blocks = 'A,B,C,D,E,F,G,H,I,J,K,L,M,N,O,P,Q'
Blocks = Blocks.split(',')
for i in range(len(_)):
    correlations.append([np.corrcoef(_[i]['y_test'],_[i]['y_hat'])[0][1]])
correlations = np.array(correlations, dtype = float)
df_Blocks = pd.DataFrame({'CACI':Blocks, 'Correlation':correlations[:,0]})
df_Blocks
```

```python
AA = df[(df.Acorn == 'A') | (df.Acorn == 'B') | (df.Acorn == 'C')]
RP = df[(df.Acorn == 'D') | (df.Acorn == 'E')]
CC = df[(df.Acorn == 'F') | (df.Acorn == 'G') | (df.Acorn == 'H') | (df.Acorn == 'I') | (df.Acorn == 'J')]
FS = df[(df.Acorn == 'K') | (df.Acorn == 'L') | (df.Acorn == 'M') | (df.Acorn == 'N')]
UA = df[(df.Acorn == 'O') | (df.Acorn == 'P') | (df.Acorn == 'Q')]
```

```python
CACI = [AA, RP, CC, FS, UA]
abbrev = 'AA, RP, CC, FS, UA'
abbrev = abbrev.split(', ')
caci_cor = []
for i in range(len(CACI)):
    caci_cor.append([np.corrcoef(CACI[i]['y_test'],CACI[i]['y_hat'])[0][1]])
caci_cor = np.array(caci_cor, dtype = float)
df_CACI = pd.DataFrame({'CACI':abbrev, 'Correlation':caci_cor[:,0]})
df_CACI.head()
```
